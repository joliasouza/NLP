```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df["reviewDescription"], df["Positively rated"], random_state=50)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
X_train[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(min_df=5, ngram_range=(1,2)).fit(X_train)
```

```python
len(vect.get_feature_names_out())
```

```python
X_train_vectorized = vect.transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_vectorized, y_train)
```

```python
#save predictions
predictions = model.predict(vect.transform(X_test))
```

```python
print(predictions)
```

```python
#auc score
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test, predictions))
```

```python
feature_names = np.array(vect.get_feature_names_out())
sorted_coef_index = model.coef_[0].argsort()
```

```python
print("Negative words", feature_names[sorted_coef_index[:10]])
```

```python
print("Positive words", feature_names[sorted_coef_index[:-11:-1]])
```
